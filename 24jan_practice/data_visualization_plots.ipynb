```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y,color='green',linestyle='dotted',linewidth=2,marker='*')
plt.title('customised line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[1000,1100,1200,1300,1400,1500]
plt.plot(x,y,color='green',linestyle='dashed',linewidth=2,marker='*')
plt.title('savings account over 6 months')
plt.xlabel('months(1-6)')
plt.ylabel('savings amount')
plt.show()
```

```python
cat=['A','B','C','D','E']
val=[10,20,15,8,14]
plt.bar(cat,val,color=['green','pink','blue','grey','skyblue'],edgecolor='black',width=0.8)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
#histogram
ages=[20,22,23,25,27,27,28,30,31,32,50,45,44,42,53,66,77]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='green',edgecolor='black',width=4)
plt.title('Age distribution histogram')
plt.xlabel('ages')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

```python
x=['python','C','Java','C++','DSA']
y=[30,40,35,48,40]
plt.pie(y,labels=x,colors=['green','pink','blue','yellow','skyblue'],autopct='%1.1f%%',startangle=90)
#'%1.1f%%' is a format specifier that gives percentage values upto one decimal values
# startangle=90,satrts of the pie chart to 90 degrees
plt.legend()          # that index box we use legend
plt.title('languages usage pie chart')
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Scatter plot")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.show()
```
