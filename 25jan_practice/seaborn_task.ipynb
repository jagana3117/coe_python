```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('test.csv')
data.head()
```

```python
train=pd.read_csv('train.csv')
train.head()
```

```python
sns.barplot(data=train,x='Pclass',y='Survived', palette={'blue',  'green', 'yellow'},hue=None)
plt.title('survival rate acress passengers')
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.show()
```

```python
sns.scatterplot(data=train,x='Age',y='Fare',palette='crest',marker='*',hue='Survived')
plt.title('age vs fare')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
subset=train[['Age','Fare','Pclass']]
sns.pairplot(subset,hue='Pclass',palette='crest')
plt.show()
```

```python
numeric_train= train.select_dtypes(include=['float64', 'int64'])
correlation_matrix=numeric_train.corr()
sns.heatmap(correlation_matrix,cmap='flare',annot=True,fmt='.2f')
plt.show()
```
