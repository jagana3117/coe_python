```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color='green')#kde for the curve on the plot
plt.title('Distribution of total bills')
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill',fill=True,color='purple')
plt.title('kde plot of total bills')
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',palette='viridis')
plt.title('Average total bill by day')
plt.show()
```

```python
sns.color_palette()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',palette='flare')
plt.title('box plot for total bill by day')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',palette='crest')
plt.title('Scatter plot of Total bill vs Tip')
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',marker='*',color='brown')
plt.title('line plot of party size vs Tip')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='husl')
plt.show()
```

```python
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_tips.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='flare',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
